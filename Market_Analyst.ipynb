# **LATAR BELAKANG**

Dalam konteks strategi pemasaran, Tim Customer Relationship Management (CRM) saat ini tengah berupaya mengembangkan sebuah model prediksi. Model ini dirancang untuk memproyeksikan bagaimana pelanggan akan merespons tawaran baru, dengan tujuan akhir meningkatkan efektivitas seluruh strategi pemasaran yang diterapkan. Sebagai langkah awal, tim CRM telah melakukan uji coba dengan menggunakan data dari 2.205 pelanggan.

Tujuan utama dari inisiatif ini adalah mengoptimalkan model prediksi guna meminimalkan kesalahan negatif. Dalam konteks ini, kesalahan negatif mengacu pada situasi di mana model tidak mampu memprediksi respons pelanggan yang sebenarnya akan merespons tawaran baru. Hal ini menjadi masalah karena dapat mengakibatkan peluang yang berharga terlewatkan, yang pada gilirannya dapat berdampak pada pendapatan perusahaan.

Sebaliknya, kesalahan positif palsu juga menjadi perhatian tim CRM. Ini terjadi ketika model memprediksi respons pelanggan padahal sebenarnya mereka tidak akan merespons tawaran tersebut. Kesalahan semacam ini akan mengakibatkan biaya pengiriman penawaran yang tidak perlu, serta risiko kehilangan peluang untuk mengirimkan tawaran yang lebih relevan kepada pelanggan yang sebenarnya berpotensi.

Hal terakhir yang perlu diperhatikan adalah keterbatasan anggaran yang dimiliki oleh tim CRM. Anggaran mereka hanya memungkinkan pengiriman penawaran kepada maksimal 35% dari total pelanggan dalam database mereka. Dengan kata lain, model yang akan dikembangkan harus mampu memberikan rekomendasi yang cerdas mengenai pelanggan mana yang sebaiknya menerima tawaran, sehingga tetap memperhatikan keterbatasan anggaran yang ada.

Keseluruhan inisiatif ini sangat penting karena pada akhirnya, tujuan utama Tim CRM adalah untuk mengoptimalkan penggunaan anggaran pemasaran mereka. Dengan cara ini, mereka dapat mengirimkan penawaran hanya kepada pelanggan yang memiliki potensi besar untuk meresponsnya, meningkatkan efisiensi pemasaran, dan pada gilirannya, menghasilkan hasil yang lebih baik bagi perusahaan.

Secara keseluruhan, tujuan dari penelitian ini adalah untuk menciptakan alat prediksi yang kuat dan cerdas yang akan meningkatkan efisiensi dan efektivitas strategi pemasaran tim CRM, sambil meminimalkan biaya dan risiko kesalahan dalam pengiriman penawaran kepada pelanggan.

## **Identifikasi Nilai Yang Janggal (Outlier)**

```python
# Identifikasi nilai negatif pada kolom mntregularprods

df[df['mntregularprods'] < 0]
```

```python
# Hapus nilai dengan mntregularprods negatif
df = df[df['mntregularprods'] >= 0]
df
```

```python
# Mengubah tipe data kolom menjadi tipe data boolean jika kolom tersebut hanya memiliki dua nilai unik yang adalah 0 dan 1
df_out = df.copy()

# Loop melalui seluruh kolom dalam DataFrame
for column in df_out.columns:
    # Memeriksa apakah kolom hanya memiliki dua nilai unik (0 dan 1)
    if len(df_out[column].unique()) == 2 and set(df_out[column].unique()) == {0, 1}:
        # Mengubah tipe data kolom menjadi boolean
        df_out[column] = df_out[column].astype(bool)

df_out.info()
```

```python
# Mengidentifikasi kolom-kolom bertipe data integer
col = df_out.select_dtypes(include=['int', 'float64']).columns

# Membuat DataFrame untuk menyimpan informasi outlier
outlier_info = pd.DataFrame(columns=['Column', 'Median + 1.5*IQR', '95th Percentile', 'Outliers Count', 'Unique Values'])

for column in col:
    q1 = df_out[column].quantile(0.25)
    q3 = df_out[column].quantile(0.75)
    q95 = df_out[column].quantile(0.95)
    iqr = q3 - q1
    median = df_out[column].median()
    outlier_limit = median + 1.5 * iqr

    # Menghitung jumlah nilai di atas batas outlier
    outlier_count = (df_out[column] > outlier_limit).sum()

    # Menambahkan baris informasi outlier ke DataFrame
    outlier_info = outlier_info.append({
        'Column': column,
        'Median + 1.5*IQR': outlier_limit,
        '95th Percentile': q95,
        'Outliers Count': outlier_count,
        'Unique Values': df[column].nunique()
    }, ignore_index=True)

# Menampilkan hasil dalam bentuk tabel
outlier_info
```

# **REGRESI LOGISTIK**

Dalam konteks analisis ini, Binomial Logistic Regression (Regresi Logistik Binomial) digunakan sebagai salah satu alat statistik yang mendasari pembangunan model prediksi. Model ini dirancang untuk memproyeksikan bagaimana pelanggan akan merespons tawaran baru, dengan fokus utama pada pengambilan keputusan pemasaran. Regresi logistik binomial sangat relevan dalam konteks ini karena masalah yang dihadapi tim Customer Relationship Management (CRM) adalah bagaimana memprediksi apakah seorang pelanggan akan "merespons" (1) atau "tidak merespons" (0) terhadap tawaran pemasaran.

```python
df_reg = df.copy()
```

```python
df_reg.columns
```

```python
df_reg = df_reg.drop(['avg_purchase_per_transaksi', 'pembelian_dengan_penawaran_khusus(%)',
                    'acceptedcmp3', 'acceptedcmp4', 'acceptedcmp5', 'acceptedcmp1', 'acceptedcmp2'
                   ], axis=1)
df_reg
```

**Outliers, atau data ekstrem, dapat berdampak buruk pada model regresi logistik, terutama dalam dataset kecil. Oleh karena itu, kita akan mengatasi outliers tersebut dengan menggantikannya dengan nilai batasan yang dihitung berdasarkan Median dan Interquartile Range. Hal ini akan membantu meningkatkan kestabilan dan akurasi model**

```python
col = ['income', 'recency', 'mntwines', 'mntfruits',
       'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts',
       'mntgoldprods', 'numdealspurchases', 'numwebpurchases',
       'numcatalogpurchases', 'numstorepurchases', 'numwebvisitsmonth',
       'age', 'customer_days','mnttotal', 'mntregularprods']

for column in col:
    q1 = df_reg[column].quantile(0.25)
    q3 = df_reg[column].quantile(0.75)
    iqr = q3 - q1
    median = df_reg[column].median()
    limit = median + 1.5*iqr

    df_reg.loc[df_reg[column] > limit, column] = limit
    print('{:>20} | Batas Atas: {}'.format(column, limit))
```

**Tabel memberikan gambaran tentang batas atas yang digunakan untuk mengatasi outliers dalam data untuk setiap variabel**

```python
# Penskalaan data sehingga memiliki mean = nol dan stdev = 1
scaler = StandardScaler()

# Penskalaan kolom tertentu dalam DataFrame
df_reg[col] = scaler.fit_transform(df_reg[col])

# Tampilkan DataFrame setelah penskalaan
df_reg
```

```python
# Logistic Regression
# Mengambil kolom 'Response' dari DataFrame df_reg dan menetapkannya ke variabel y
y = df_reg['response']

# Mengambil seluruh kolom dari DataFrame df_reg kecuali kolom 'Response' dan menetapkannya ke variabel x
X = df_reg.drop('response', axis=1)
```

```python
# Membagi data menjadi training set (80%) dan testing set (20%)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)
```

```python
# Membangun model regresi logistik dan melatihnya dengan data pelatihan (training set)
log_model = LogisticRegression(random_state=42, max_iter=2000)
reg_output = log_model.fit(X_train, y_train)
```

```python
 # Memproses data dari X_test dan menghasilkan prediksi yang akan disimpan dalam variabel y_pred
 y_pred = reg_output.predict(X_test)
 y_pred
```

```python
# mengukur tingkat akurasi model regresi logistik

results = [
    ["Accuracy", "%.3f" % accuracy_score(y_test, y_pred)],
    ["Precision", "%.3f" % precision_score(y_test, y_pred)],
    ["Recall", "%.3f" % recall_score(y_test, y_pred)],
    ["F1 Score", "%.3f" % f1_score(y_test, y_pred)]
]

table = tabulate(results, headers=["Metric", "Score"], tablefmt="pretty")
print(table)
```

```python
# Menghitung confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Membuat visualisasi confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=reg_output.classes_)

# Plot matriks kebingungan
disp.plot(values_format='d')

plt.show()

# Hasilkan tabel matriks kebingungan
results = [
    ["True Positives", cm[1, 1]],
    ["True Negatives", cm[0, 0]],
    ["False Positives", cm[0, 1]],
    ["False Negatives", cm[1, 0]]
]

table = tabulate(results, headers=["Metric", "Value"], tablefmt="pretty")
print(table)
```

**Kesimpulan :**

* Recall adalah metrik yang mengukur sejauh mana model mampu mengidentifikasi seluruh respons positif yang benar-benar ada dalam data. Ini penting karena recall menggambarkan kemampuan model dalam "menangkap" pelanggan yang sebenarnya akan merespons tawaran baru.  Dalam konteks dataset ini, nilai recall sekitar 0.500 atau 50%. Ini berarti bahwa **model hanya berhasil mengidentifikasi sekitar 50% dari pelanggan yang seharusnya merespons tawaran**

* Presisi adalah metrik yang mengukur seberapa baik model mampu memprediksi respons positif dengan benar. Dalam konteks ini, presisi sekitar 0.738 atau 73.8%. Ini berarti dari **pelanggan yang diprediksi oleh model akan merespons, sekitar 73.8% dari mereka benar-benar merespons tawaran**

```python
# Mengidentifikasi fitur-fitur yang paling berpengaruh dalam memprediksi respons pelanggan terhadap tawaran baru

feature_importance = sorted(list(zip(X_train.columns, log_model.coef_[0])), key=lambda x: x[1], reverse=True)
feature_importance
```

```python
# Visualisasikan data

features, coefficients = zip(*feature_importance)

# Create a bar chart
plt.figure(figsize=(10, 5))
plt.barh(features, coefficients, color='b')
plt.xlabel('Coefficient Value')
plt.title('Fitur-Fitur Penting Dalam Model Regresi Logistik')
plt.gca().invert_yaxis()
plt.show()
```

**Kesimpulan :**

* Fitur **acceptedcmpoverall** memiliki koefisien tertinggi sekitar **1.721**, yang berarti fitur ini memiliki pengaruh positif yang signifikan terhadap pelanggan yang merespons tawaran. Kemungkinan besar, fitur ini mengacu pada hasil kampanye penerimaan secara keseluruhan, yang cenderung meningkatkan respons

* Fitur **education_phd** memiliki koefisien positif tinggi sekitar **0.926**. Pelanggan dengan **gelar doktor (Ph.D.) cenderung merespons tawaran dengan baik**

* Fitur **customer_days** memiliki koefisien positif sekitar **0.742**, menunjukkan bahwa **pelanggan yang telah menjadi pelanggan untuk waktu yang lama lebih mungkin merespons tawaran baru**

Recall adalah kemampuan model untuk mengidentifikasi sebanyak mungkin pelanggan yang benar-benar akan merespons tawaran. Namun, saat ini tingkat recall masih rendah, yang berarti model tidak dapat mengidentifikasi sebagian besar pelanggan yang seharusnya merespons.

Untuk meningkatkan recall, tim kami memutuskan untuk mengubah ambang keputusan model (decision threshold). Model regresi logistik memiliki ambang keputusan default sebesar 0.5. Artinya, pelanggan dengan probabilitas respons kurang dari 50% diprediksi tidak akan merespons, sementara yang di atas 50% akan merespons.

Namun, tim memutuskan untuk menurunkan ambang ini menjadi 30%. Artinya, pelanggan dengan probabilitas respons di atas 30% akan diprediksi sebagai pelanggan yang akan merespons. Ini akan meningkatkan jumlah pelanggan yang dianggap sebagai calon responden, tetapi juga akan mengakibatkan lebih banyak "false positives," yaitu pelanggan yang diprediksi akan merespons tetapi sebenarnya tidak melakukannya.

```python
# mendapatkan nilai probabilitas dari hasil prediksi model regresi logistik (404 data hasil prediksi)
pred_probabilitas = reg_output.predict_proba(X_test)

# menciptakan sebuah daftar yang hanya berisi semua nilai probabilitas terkait dengan pelanggan yang akan merespons
prob = [x[1] for x in pred_probabilitas]

# Jika probabilitas respons pelanggan (x) >= 0.3 maka dikategorikan pelanggan merespon, jika < 0.3, berarti pelanggan tidak akan merespons (0)
y_pred_new = np.array([1 if x >= 0.3 else 0 for x in prob])
y_pred_new
```

```python
# mengukur tingkat akurasi model regresi logistik

results = [
    ["Accuracy", "%.3f" % accuracy_score(y_test, y_pred)],
    ["Precision", "%.3f" % precision_score(y_test, y_pred)],
    ["Recall", "%.3f" % recall_score(y_test, y_pred)],
    ["F1 Score", "%.3f" % f1_score(y_test, y_pred)]
]

table = tabulate(results, headers=["Metric", "Score"], tablefmt="pretty")
print(table)
```

```python
# mengukur tingkat akurasi model regresi logistik
results = [
    ["Accuracy", "%.3f" % accuracy_score(y_test, y_pred_new)],
    ["Precision", "%.3f" % precision_score(y_test, y_pred_new)],
    ["Recall", "%.3f" % recall_score(y_test, y_pred_new)],
    ["F1 Score", "%.3f" % f1_score(y_test, y_pred_new)]
]

table = tabulate(results, headers=["Metric", "Score"], tablefmt="pretty")
print(table)
```

**Kesimpulan :**

Tingkat recall **meningkat dari 50% menjadi 59.7%** dengan menurunkan ambang probabilitas menjadi **30%**. Namun, tingkat presisi menurun dari **73.8% menjadi 57.8%**

Hal terakhir yang perlu diperhatikan adalah keterbatasan anggaran yang dimiliki oleh tim CRM. Anggaran mereka hanya memungkinkan pengiriman penawaran kepada maksimal 35% dari total pelanggan dalam database mereka. Dengan kata lain, model yang akan dikembangkan harus mampu memberikan rekomendasi yang cerdas mengenai pelanggan mana yang sebaiknya menerima tawaran, sehingga tetap memperhatikan keterbatasan anggaran yang ada.

```python
round(y_pred_new.mean(),3)
```

**Kesimpulan :**

Setelah mengubah ambang probabilitas prediksi, hasil prediksi kami menunjukkan bahwa sekitar **16% dari pelanggan diprediksi akan merespons tawaran**. Hal ini berarti tingkat respons yang diprediksi **lebih rendah dari target 35% yang diinginkan oleh tim CRM**

```python
prob = [x[1] for x in pred_probabilitas]
prob
```

```python
# memuat list daftar ambang probabilitas yang akan diuji
ambang_prob_list = np.arange(0.3, 0.01, -0.01)

# menyimpan ambang probabilitas yang ideal jika ditemukan, yaitu tingkat respon dengan target 35% pelanggan
prob_target_35 = 0

# melakukan iterasi melalui possible_probs
for list_prob in ambang_prob_list:
  # daftar prediksi yang diperoleh dengan mengambil probabilitas dari probs yang lebih besar atau sama dengan possible_prob
  preds = np.array([1 if x >= list_prob else 0 for x in prob])
  if preds.mean() > 0.35:
        prob_target_35 = list_prob
        break

y_pred_new_new = np.array([1 if x >= prob_target_35 else 0 for x in prob])

prob_target_35
```

**Kesimpulan :**

Dengan menggunakan ambang probabilitas sekitar **0.09**, model akan memprediksi respons dari sekitar **35% pelanggan**, sesuai dengan target yang diinginkan oleh tim CRM. Dengan kata lain, ambang probabilitas ini memungkinkan untuk **mencapai tingkat respons yang diharapkan tanpa melebihi batasan anggaran yang tersedia**

```python
# mengukur tingkat akurasi model regresi logistik
results = [
    ["Accuracy", "%.3f" % accuracy_score(y_test, y_pred_new_new)],
    ["Precision", "%.3f" % precision_score(y_test, y_pred_new_new)],
    ["Recall", "%.3f" % recall_score(y_test, y_pred_new_new)],
    ["F1 Score", "%.3f" % f1_score(y_test, y_pred_new_new)],
    [],
    ["Total Pelanggan Yang Merespon Penawaran Perusahaan", "%.3f" % y_pred_new_new.mean()]
]

table = tabulate(results, headers=["Metric", "Score"], tablefmt="pretty")
print(table)
```

**Kesimpulan :**

Dengan menggunakan ambang probabilitas sebesar 0.11 untuk memprediksi respons pelanggan terkait kampanye yang diberikan :

* Tingkat **recall** meningkat menjadi **90.3%** yang berarti model mampu mengidentifikasi sebagian besar pelanggan yang seharusnya merespons tawaran **dengan sangat baik**

* **Presisi** menurun menjadi **40.3%**, yang berarti model lebih cenderung **memprediksi lebih banyak respons positif**, tetapi sebagian di antaranya **sebetulnya adalah false positive (pelanggan yang sebenarnya tidak akan merespons)**

* **Model akan memprediksi bahwa sekitar 35.6% dari pelanggan akan merespons tawaran yang sesuai dengan anggaran yang dimiliki oleh tim CRM**
